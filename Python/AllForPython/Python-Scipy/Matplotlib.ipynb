* Matplotlib  
    Matplotlib是一个最常用的2D绘图包，可以完成绝大部分的绘图任务，当然也可以完成部分简单的3D绘图任务，效果并不如其他的库那么理想

```python
# 获取数据
%matplotlib inline
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.cos(x), np.sin(x)
plt.plot(x, c, x, s)
```

```python
plt.figure(figsize=(8, 6), dpi=80)    # (8, 6) 英寸，每英寸2,54cm, 分辨率是80
plt.subplot(1, 1, 1)    # 传建一个子图 1 * 1,占据第1个网格
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.cos(x), np.sin(x)
# color 设置颜色，设置线宽，设置线的样式,alpha还可以设置我们的透明度
plt.plot(x, c, color='blue', linewidth=1.0, linestyle='-')
plt.plot(x, s, color='green', linewidth=1.0, linestyle='-')
# 设置x轴的范围
plt.xlim(-4.0, 4.0)
# 设置轴的说明
plt.xlabel('这是x')
# 设置坐标轴的刻度,支持使用Latex标注
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.ylim(-1.0, 1.0)
plt.ylabel('这是y')
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

ax = plt.gca()

# 为了可以让我们的刻度显示出来不被线条遮盖
for label in ax.get_xticklabels() + ax.get_yticklabels():
    # 将刻度的字体大小调整
    label.set_fontsize(16)
    # 字体的矩形显示框
    label.set_bbox(dict(facecolor='blue', edgecolor='None', alpha=0.65))


plt.show()

# plt.savefig('filename.png', dpi=80)    # 保存分辨率是８０的图图片到对应的路径上
```

```python
# spines的移动
ax = plt.gca()    # 获取当前的坐标轴
# 单个运行下面的语句，将右边的边界置空
ax.spines['right'].set_color('none')     # right, top, left, bottom
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')    # 将xaxis， x坐标轴设定成底边的bottom spines
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', 0.5))    # 将位置设定在y轴的0.5上
ax.spines['left'].set_position(('data', 0.5))    # 将位置设定在x轴的0.5刻度上
plt.show()
```

```python
# 添加标签
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.cos(x), np.sin(x)
plt.plot(x, c, color='blue', linewidth=2.5, linestyle='-', label='cosline')    # label属性制定之后的legend的内容
plt.plot(x, s, color='red', linewidth=2.5, linestyle='-', label='sine')
plt.legend(loc='upper left')    # 将lengend图标显示在左上角，不加的话，默认将其添加在最合适的位置上
```

```python
# 注释 annotate
ax = plt.gca()    # 获取当前的坐标轴
# 单个运行下面的语句，将右边的边界置空
ax.spines['right'].set_color('none')     # right, top, left, bottom
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')    # 将xaxis， x坐标轴设定成底边的bottom spines
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', 0))    # 将位置设定在y轴的0.5上
ax.spines['left'].set_position(('data', 0))    # 将位置设定在x轴的0.5刻度上

# 为了可以让我们的刻度显示出来不被线条遮盖
for label in ax.get_xticklabels() + ax.get_yticklabels():
    # 将刻度的字体大小调整
    label.set_fontsize(16)
    # 字体的矩形显示框
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65))

# 注释的代码
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.cos(x), np.sin(x)
plt.plot(x, c, color='blue', linewidth=10, linestyle='-', label='cosline', alpha=0.5)    # label属性制定之后的legend的内容
plt.plot(x, s, color='red', linewidth=2.5, linestyle='-', label='sine')
plt.legend(loc='upper left')    # 将lengend图标显示在左上角，不加的话，默认将其添加在最合适的位置上

t = 2 * np.pi / 3
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=2.5, linestyle='--')    # 显示直线(t,0)->(t,cos(t))
plt.scatter([t, ], [np.cos(t), ], 50, color='blue')    # 将(t,np.cos(t))处打上点，显示

# 第一个参数是注释的文本Latex嵌入，注释的目标点是(t, cos(t)), 注释的文本坐标是(-90, -100), 距离使用offset point相对与终点的
# arrowprops箭头的属性 , arrowstyle是箭头的格式，还有连接的方式
plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', \
             xy=(t, np.cos(t)), xycoords='data', xytext=(-90, -100), \
             textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))

t = 2 * np.pi / 3
plt.plot([t, t], [0, np.sin(t)], color='blue', linewidth=2.5, linestyle='--')    # 显示直线(t,0)->(t,cos(t))
plt.scatter([t, ], [np.sin(t), ], 50, color='blue')    # 将(t,np.cos(t))处打上点，显示

plt.annotate(r'$sin(\frac{2\pi}{3})=-\frac{\sqrt{3}}{2}$', \
             xy=(t, np.sin(t)), xycoords='data', xytext=(-90, -100), \
             textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))
```

* 数字，子图，轴的技巧使用
    * figure : 是一个窗口
    * subplots : 是一个窗口的子图,默认使用figure创建的是subplots(111)

```python
%matplotlib inline
# num是图的编号，不指定默认自增，figsize是图片的边框大小，facecolor是前景色，edgecolor是后景色，frameon是是否绘制图框
plt.figure(num=1, figsize=(80, 80), dpi=100, facecolor='green', edgecolor='blue', frameon=True)
plt.show()
plt.close()
```

```python
# 颜色填充
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y= np.sin(2 * X)

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.plot(X, Y + 1, color='blue', alpha=1.00)
plt.fill_between(X, 1, Y + 1, color='blue', alpha=0.25)    # 1 - Y + 1之间使用蓝色填充

plt.plot(X, Y - 1, color='blue', alpha=1.00)
plt.fill_between(X, -1, Y - 1, (Y - 1) > -1, color='blue', alpha=0.25)    # where数组选择性填充
plt.fill_between(X, -1, Y - 1, (Y - 1) < -1, color='red', alpha=0.25)
```

```python
# 散点图
n = 1024
X = np.random.normal(0, 1, n)    # 高斯分布,均值，标准差，规模
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X)

plt.axes([0.25, 2.5, 0.95, 0.95])
plt.scatter(X, Y, s=75, c=T, alpha=0.5)    # size大小，c颜色，alpha是透明度
```

```python
# 条形图
n = 12
x = np.arange(n)
y1 = (1 - x / float(n)) * np.random.uniform(0.5, 1.0, n)    # 均匀分布中随机采样,[low, high),size是规模
y2 = (1 - x / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(x, +y1, facecolor='#9999ff', edgecolor='white')    # 条形图
plt.bar(x, -y2, facecolor='#ff9999', edgecolor='white')

for x1, y in zip(x, y1):
    plt.text(x1, y + 0.05, '%.2f' % y)    # x,y是文本注释的坐标,第三个参数显示的内容,alpha是透明度,family设置字体
for x1, y in zip(x, y2):
    plt.text(x1 + 0.4, - y - 0.05, '%.2f' % y, ha='center', va='top')
```

```python
# 等高线
# np.meshgrid函数用来用两个一维数组(x, y)生成两个二维网格(X, Y)
# 其中X是x的简单复制，Y是y的简单复制,两个矩阵的规模都是n * m(x - m, y - n)
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(- x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.imshow(Z, interpolation='nearest', cmap='bone', origin='lower')
plt.colorbar(shrink=0.92)
```

```python
# 饼图
n = 20
Z = np.ones(n)
plt.axes([0.025, 0.025, 0.95, 0.95])

plt.pie(Z, explode=Z*0.05, colors=['%f' % (i / float(n)) for i in range(n)])
plt.axis('equal')   # 刻度对齐，保证是标准圆形
plt.xticks()
plt.yticks()

plt.show()
```
