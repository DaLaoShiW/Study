* 网络原始文本获取并进行内容解析

```python
from __future__ import division
import nltk, re, pprint, requests
from bs4 import BeautifulSoup
```

```python
url = 'http://www.gutenberg.org/files/2544/2544.txt'
ans = requests.get(url)
ans.encoding = 'utf8'
print(type(ans.text), len(ans.text))    # 获取文本属性
print(ans.text)
```

```python
url2 = 'http://news.bbc.co.uk/2/hi/health/2284783.stm'
html = requests.get(url2)
html.encoding = 'utf8'
soup = BeautifulSoup(html.text, 'lxml')
clean = soup.get_text()
c = nltk.Text(clean)
print(c.concordance('gene'))
```

* NLP处理流程
    * 数据文件的获取
    * 文本信息抽取和分词
        尽量将处理的分词结果存放在list等容器中
    * 标准化处理

```python
#i 英文分词
a = 'i am the one who can get everything for me！'
nltk.word_tokenize(a)    # 英文分词调用
```

```python
# 词干抽取
raw = '''DENNIS: Listen, strange women lying in ponds distributing swords
         is no basic for a system of government. Supreme executive power derives from a mandate from the masses
         , not from som farcical aquatic ceremony, do, did, done'''
tokens = nltk.word_tokenize(raw)
print(tokens)
# 词干提取器
# 虽然正则表达式可以完成这样的功能，但是为了正确性和稳定性期间，强烈建议使用nltk提供的词干提取器
# 一般使用porter磁感提取器就可以达到很好的效果
porter = nltk.PorterStemmer()    # porter词干提取器
lancaster = nltk.LancasterStemmer()    # lancaster词干提取器
print([porter.stem(t) for t in tokens])    # 该实例中,lying成功的被porter词干提取器提取出来，但是ｌａｎｃａｓｔｅｒ失败了
print([lancaster.stem(t) for t in tokens])

# nltk.Index函数的处理
# nltk.Index函数的输入是二元元组的列表，元组的0号元素是键，1号元素是值，生成对应的Index对象，每一个键回应涉一个list其中包含满足的键值对的值属性
index = nltk.Index([(porter.stem(word), i) for i, word in enumerate(tokens)])
print(index)
```

```python
# 词形转换预处理, 将do, did, done（例子如上）统一转化成do预处理,内部使用字典来实现转换工作,转换的速度会比较慢
wnl = nltk.WordNetLemmatizer()    # 词型转换器
print([wnl.lemmatize(t) for t in tokens])
```
