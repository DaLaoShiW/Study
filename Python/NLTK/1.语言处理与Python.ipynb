```python
import nltk
nltk.__version__
```

```python
# 加载book文本,语料
from nltk.book import *
```

```python
text1.concordance('monstrous')    # 查找nltk.book中出现的单词并显示出对应的上下文文本
```

```python
text1.similar('monstrous')    # 和monstrous出现在相似上下文的单词
```

```python
# 基本操作
print(len(text1))    # 文本计数,单词和标点符号（合称标识符）
# print(sorted(set(text1)))     # 集合排序，去除相同的单词，只保留一个
print(list(text1).count('the'))    # 统计出现过多少个the
```

```python
# 简单文本统计,将文本名传递作为参数
fdist = FreqDist(text1)     # 统计对象，提供类似字典的API,键是词，值是出现的次数(可以理解成是频度)，只能接收词向量
print(fdist)
print(fdist['the'])    # 输出结果和上面的匹配
print(fdist.freq('the'))    # 输出单词样本的频率
print(fdist.N())    # 输出样本总数
print(fdist.max())    # 输出频率最大样本单词
print(len(fdist.keys()))   # 输出键
# freqdist对象可以用来进行 >, <, == 的判断操作，但是必须是所有的成员都满足的情况才会对应的返回True 
```

```python
# 图示
%matplotlib inline
fdist.plot(50, cumulative=True)     # cumulative是累计求和，可以看出词数的增长情况,现实前50最大频度的单词
fdist.plot(10, cumulative=False)
```

```python
# 低频次的统计查看结果
len(fdist.hapaxes())    # 统计出现过一次的单词
# 如果想要统计出现过固定单词的个数，可以考虑对整个FreqDist字典做枚举（列表推倒式）
```

```python
# 词语搭配，双连词,之后的条件频率分布有用
# 三连词 : nltk.trigrams(list...)
for i in nltk.bigrams(['there', 'is', 'a', 'rabbit', 'on', 'my', 'head']):
    print(i)
text1.collocations()    # 扇面的双连词非常的不准确，我们最常使用的双连词就需要使用collocations来实现
```
